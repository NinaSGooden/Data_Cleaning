```python
churn_clean = pd.read_csv('C:/Users/nsgoo/OneDrive/Documents/School/WGU/d206_data_cleaning/hfy4k43ojf892f0dy7sd/churn_updated_data_V2.csv')
```

```python
data = churn_clean.loc[:, 'Tenure':'Evidence of Active Listening']
data.head()
```

```python
from sklearn.decomposition import PCA
```

```python
churn_normalized = (data - data.mean()) / data.std()
```

```python
pca = PCA(n_components = data.shape[1])
```

```python
churn_numeric = data[['Tenure', 'MonthlyCharge', 'Bandwidth_GB_Year','Timely Response', 'Timely Fixes', 'Timely Replacements','Reliability', 'Options', 'Respectful Response', 'Courteous Exchange', 'Evidence of Active Listening']]
pca_names = []
for i, col in enumerate(churn_numeric.columns):
    pca_names.append('PC' + str(i + 1))
print(pca_names)
```

```python
pca.fit(churn_normalized)
churn_pca = pd.DataFrame(pca.transform(churn_normalized),
columns = pca_names)
```

```python
plt.plot(pca.explained_variance_ratio_)
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance')
plt.show();
```

```python
cov_matrix = npy.dot(churn_normalized.T, churn_normalized) / data.shape[0]
eigenvalues = [npy.dot(eigenvector.T, npy.dot(cov_matrix, eigenvector)) for
    eigenvector in pca.components_]
```

```python
plt.plot(eigenvalues)
plt.xlabel('Number of Components')
plt.ylabel('Eigenvalue')
plt.show();
```

```python
for pc, var in zip(pca_names, npy.cumsum(pca.explained_variance_ratio_)):
    print(pc, var)
```

```python
churn_rotation = pd.DataFrame(pca.components_.T, columns = pca_names, index = churn_numeric.columns)
print(churn_rotation)
```
